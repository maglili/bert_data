# set0

## train-test split

```
import pandas as pd
import numpy as np
```

```
og_df = pd.read_pickle('/content/drive/MyDrive/Project_data/BERT/embedding_table/neg_set_test/output_table0.pkl')
og_df
```

```
X = og_df['word_embedding']
X
```

```
y = og_df['word_label']
y
```

### Data spliting to train / dev

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=0)
```

```
X_train
```

```
print("Train data:", len(X_train)) 
print("Test data:", len(X_test))
```

```
print('label = 1(Pos) in train: ', len(y_train[y_train ==1]))
print('label = 0(Neg) in train: ', len(y_train[y_train ==0]))
```

```
X_test
```

```
print('label = 1(POS) in test: ', len(y_test[y_test ==1]))
print('label = 0(NEG) in test: ', len(y_test[y_test ==0]))
```

## N-fold

```
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=0)
```

```
skf.get_n_splits(X_train, y_train)
```

```
train_cv = {}

for i, (train_index, test_index) in enumerate(skf.split(X_train, y_train)):
  train_cv['train'+str(i)] = pd.DataFrame({'word_embedding': X_train.iloc[train_index],'word_label': y_train.iloc[train_index]} )
  train_cv['cv'+str(i)] = pd.DataFrame({'word_embedding': X_train.iloc[test_index],'word_label': y_train.iloc[test_index]} )
  a = y_train.iloc[train_index]
  b = y_train.iloc[test_index]
  print('train:')
  print(train_cv['train'+str(i)])
  print('valid:')
  print(train_cv['cv'+str(i)])
  print('label = 1(POS) in train: ',len(a[a == 1]))
  print('label = 0(NEG) in train: ',len(a[a == 0]))
  print('label = 1(POS) in valid: ',len(b[b == 1]))
  print('label = 0(NEG) in valid: ',len(b[b == 0]))
  print('-'*80)
```

## save as pickle files

```
train0 = train_cv['train0']
train0.to_pickle("/content/drive/MyDrive/Project_data/BERT/process_data/bert_base/net_set_test/set0/train0.pkl")
train0
```

```
cv0 = train_cv['cv0']
cv0.to_pickle("/content/drive/MyDrive/Project_data/BERT/process_data/bert_base/net_set_test/set0/cv0.pkl")
cv0
```
