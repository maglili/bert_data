# split_pos

Split positive data into training / validation / test,

using sklearn **StratifiedKFold**.

The distribution is 8:1:1.

## Load data

```python
import numpy as np
import random 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import StratifiedKFold
plt.rcParams["figure.figsize"] = [12, 8]

seed_val = 0
random.seed(seed_val)
np.random.seed(seed_val)
```

```python
def year_distri(year_list:list)->dict:
    """
    Count year districution.

    Input:
        year_list(list)

    Output:
        year_count(dict): key:years, value:count
        years_X(list): Years
        years_y(list): count
    """

    year_count = {}
    for i in year_list:
        year_count[i] = year_count.get(i,0)+1

    years_X = [] # keep order
    years_y = []

    for year, count in year_count.items():
        years_X.append(int(year))
    years_X.sort()

    for year in years_X:
        years_y.append(year_count[str(year)])

    return years_X, years_y, year_count
```

```python
df = pd.read_csv('../csv_data/output-pos-remove_no_abst.csv', encoding='utf-8', header=None, dtype=str).fillna('')
df.columns = ['pmid','title','abstract','doi','year','authors']
df
```

```python
years_X, years_y, year_count = year_distri(df['year'].to_list())
```

```python
plt.bar(years_X,years_y)
plt.title('Year distribution')
plt.ylabel('count')
plt.xlabel('year')
plt.savefig('./pics/pos_distri.png')
plt.show()
```

## Shuffle DataFrame

Shuffle the dataframe by sample() method.

```python
sh_df = df.sample(frac=1, random_state=0).reset_index(drop=True)
sh_df
```

```python
X = sh_df[['pmid','title','abstract','doi','authors']].values
y = sh_df['year'].values
```

## StratifiedKFold

### Experiment

Check that sum of test index is equal to original data

```python
skf = StratifiedKFold(n_splits=10, random_state=0, shuffle=True)
y_list = []
for train_index, test_index in skf.split(X, y):
    y_list.extend(test_index)
y_list.sort()

# check
assert y_list == [x for x in range(len(y_list))]
```

```python
skf = StratifiedKFold(n_splits=10, random_state=0, shuffle=True)

index_list = []
for idx, (train_index, test_index) in enumerate(skf.split(X, y)):
    
    index_list.append(test_index)
    
    years_X, years_y, _ = year_distri(y[test_index].tolist())
    plt.bar(years_X,years_y)
    plt.title('Year distribution of Fold '+str(idx))
    plt.xlabel('year')
    plt.ylabel('count')
    axes = plt.gca()
    axes.set_xlim([1984,2021])
    axes.set_ylim([0,6])
    plt.savefig('./pics/fold-'+str(idx)+'.png')
    plt.show()
```

## Output

```python
def rmv_blank(df:pd.DataFrame):
    """
    Remove redundant blank in abstract.
    
    Args:
        df(DataFrame)
    """
    for idx in range(len(df)):
        bef_abs = df.iloc[idx]['abstract']
        bef_abs_split = bef_abs.split()
        aft_abs = ' '.join(bef_abs_split)
        df.iloc[idx]['abstract'] = aft_abs
```

### test

```python
test_df = sh_df.iloc[index_list[0]]
rmv_blank(test_df)
test_df.to_csv('../csv_data/test.csv', index=False, encoding='utf-8')
test_df
```

### validation

```python
valid_df = sh_df.iloc[index_list[1]]
rmv_blank(valid_df)
valid_df.to_csv('../csv_data/valid.csv', index=False, encoding='utf-8')
valid_df
```

### train

```python
train_index = []
for i in range(2, len(index_list)):
    train_index.extend(index_list[i])
len(train_index)
```

```python
train_df = sh_df.iloc[train_index]
rmv_blank(train_df)
train_df.to_csv('../csv_data/train.csv', index=False, encoding='utf-8')
train_df
```
