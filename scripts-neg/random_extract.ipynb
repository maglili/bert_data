# Extract negtive data
Extract from new negtive set

```python
import pandas as pd
import matplotlib.pyplot as plt

import random
random.seed(0)
import numpy as np
np.random.seed(0)

from tqdm.notebook import tqdm
```

```python
def check_year_distr(year_list: list) -> dict:
    """
    input: a list containing year.
        -e.g. year_list = [1998, 1998, 1999, 1999,...]
        
    output: a dictionary.
        -key: years
        -values: year count
    """
    year_set = {}
    for year in year_list:
        if year not in year_set:
            year_set[year] = 1
        else:
            year_set[year] += 1
    return year_set
```

## Load data

### positive data

```python
# pos data
pos_df = pd.read_csv('../csv_data/output-pos-remove_no_abst.csv', squeeze=True, header=None, encoding='utf-8', dtype=str)
pos_df = pos_df.fillna('')
pos_df
```

```python
pos_pmid = pos_df[0].tolist()

# pos year distribution
pos_year_list = pos_df[4].to_list()
pos_year_set = check_year_distr(pos_year_list)
```

### negtive data

```python
# neg data
neg_df = pd.read_csv('./neg_all-combine.csv', squeeze=True, header=None, encoding='utf-8', dtype=str)
neg_df = neg_df.fillna('')
neg_df
```

Remove review paper

If paper is review paper, then the year column will be empty.

```python
neg_df = neg_df.drop(neg_df[neg_df[4]==''].index)
neg_df
```

```python
pmid_err = 0
abst_err = 0
remove_list = []

for i in tqdm(range(len(neg_df))):
    pmid = neg_df.iloc[i][1]
    abst = neg_df.iloc[i][3]
    if pmid in pos_pmid:
        pmid_err += 1
        remove_list.append(i)
    if (abst == 'No abstract available') or (abst == ''):
        abst_err += 1
        remove_list.append(i)

remove_list = list(set(remove_list))
print('remove samples:',pmid_err+abst_err)
print('pmid error:',pmid_err)
print('abst error:',abst_err)

neg_df = neg_df.drop(remove_list)
neg_df
```

```python
# neg year distribution
neg_year_list = neg_df[4].to_list()
neg_year_set = check_year_distr(neg_year_list)
neg_year_set
```

Find Dataframe accroding to year

```python
# save all data by its published year, balance data to fit pos data
all_set = dict()

for i in range(1985,2021): # 1985 ~ 2020 
    
    all_set[str(i)] = neg_df[neg_df[4]==str(i)]
    
# extract data
num_extract = 6
extract_pmid = []
extract_result = {} # save all dataset
random_seed = 0

for i in range(num_extract):   
    
    extract_frames = []
    
    for year, appear_times in pos_year_set.items():
        
        extract_data = all_set[year].sample(n=appear_times, random_state = random_seed)     
        all_set[year] = all_set[year].drop(extract_data.index)   
        extract_frames.append(extract_data)
       
    result = pd.concat(extract_frames) #combine dataframes in frames   
    extract_result['dataset_'+str(i)] = result
    filename='neg_set'+str(i)+'.csv'
    result.to_csv('./csv_data/'+filename, header = False, index = False, encoding = 'utf-8')
```
