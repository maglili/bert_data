# Extract negtive data in balance way

Extract negative data from neg_all-plus.csv

The number of positive and negative are *queal*.

neg_all-plus has been processed by:

1. remove no year data
2. fix data year in [1985, 2020]

```python
import pandas as pd
import matplotlib.pyplot as plt
import random
import numpy as np
from tqdm.notebook import tqdm
np.random.seed(0)
random.seed(0)
```

```python
def check_year_distr(year_list: list) -> dict:
    """
    input: a list containing year.
        -e.g. year_list = [1998, 1998, 1999, 1999,...]
        
    output: a dictionary.
        -key: years
        -values: year count
    """
    year_set = {}
    for year in year_list:
        if year not in year_set:
            year_set[year] = 1
        else:
            year_set[year] += 1
    return year_set
```

## Load data

```python
# pos data
train_df = pd.read_csv('../csv_data/balance/train.csv', squeeze=True, encoding='utf-8', dtype=str)
train_df = train_df.fillna('')
train_df
```

```python
pos_pmid = train_df['pmid'].tolist()

# pos year distribution
pos_year_list = train_df['year'].to_list()
pos_year_set = check_year_distr(pos_year_list)
```

```python
neg_df = pd.read_csv('../csv_data/neg_all-plus.csv', encoding='utf-8', dtype=str)
neg_df
```

#### Processing negative data

remove:

1. positive data appear in negative
2. abstract = ''
3. avstract = 'No abstract avalible'

```python
pmid_err = 0
abst_err = 0
remove_list = []

for i in tqdm(range(len(neg_df))):
    pmid = neg_df.iloc[i]['pmid']
    abst = neg_df.iloc[i]['abstract']
    if pmid in pos_pmid:
        pmid_err += 1
        remove_list.append(i)
    if (abst == 'No abstract available') or (abst == ''):
        abst_err += 1
        remove_list.append(i)

remove_list = list(set(remove_list))
print('remove samples:', pmid_err+abst_err)
print('pmid error:', pmid_err)
print('abst error:', abst_err)

neg_df = neg_df.drop(remove_list)
neg_df
```

```python
# neg year distribution
neg_year_list = neg_df['year'].to_list()
neg_year_set = check_year_distr(neg_year_list)
neg_year_set
```
