```python
import pandas as pd
```

```python
df = pd.read_csv('../csv_data/neg_all-combine.csv', encoding='utf-8', dtype=str, header=None)
df.columns = ['title','pmid','doi','abstract','year','authors']
df = df.fillna('')
df
```

```python
remove_idx = []

for idx in range(len(df)):
    ori_abs = df.iloc[idx]['abstract']
    ori_spl = ori_abs.split()
    aft_abs = ' '.join(ori_spl)
    
    if not (ori_abs == aft_abs):
        print('error:',idx)
        
    if not (df.iloc[idx]['year']).isnumeric():
        remove_idx.append(idx)  
    elif (int(df.iloc[idx]['year'])>2020) or (int(df.iloc[idx]['year'])<1985):
        remove_idx.append(idx)
```

```python
print('len of remove_idx:', len(remove_idx))
remove_idx = list(set(remove_idx))
print('len of remove_idx:', len(remove_idx))
```

```python
df = df.drop(remove_idx)
df
```

```python
df.to_csv('../csv_data/neg_all-plus.csv', encoding='utf-8', index=False)
```
