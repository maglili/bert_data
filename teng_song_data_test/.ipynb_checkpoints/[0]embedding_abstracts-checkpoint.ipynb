# embedding_abstracts

## Part 1: Processing data

```python
import pandas as pd
import numpy as np
from tqdm.notebook import tqdm 
```

### pos

```python
df_pos = pd.read_csv('../csv_data/output-pos-remove_no_abst.csv', header = None)
df_pos = df_pos.fillna('')
df_pos
```

Extracting "abstract" column and add "label" column 

```python
df_pos = df_pos[[2]]
df_pos.columns = ['abstract']
df_pos = df_pos.assign(label = 1)
df_pos
```

Removing redundant whitespace in abstract

```python
for i in range(len(df_pos)):
    ori_text = df_pos.iloc[i][0]
    text = ori_text
    text = text.strip()
    text = text.split()
    text = ' '.join(text)
    df_pos.loc[ df_pos.abstract == ori_text, "abstract" ] = text
```

```python
df_pos
```

### neg

```python
df_neg = pd.read_csv('./neg_set4.csv', header = None)
df_neg = df_neg.fillna('')
df_neg
```

```python
df_neg = df_neg[[1]]
df_neg.columns = ['abstract']
df_neg = df_neg.assign(label = 0)
df_neg
```

```python
for i in range(len(df_neg)):
    ori_text = df_neg.iloc[i][0]
    text = ori_text
    text = text.strip()
    text = text.split()
    text = ' '.join(text)
    df_neg.loc[ df_neg.abstract == ori_text, "abstract" ] = text
```

```python
df_neg
```

### combine pos and neg dataset

```python
alldata = pd.concat([df_pos, df_neg], ignore_index=True)
alldata
```

## Part 2: Embedding

```python
import torch
from transformers import BertTokenizer, BertModel

tokenizer = BertTokenizer.from_pretrained('bert-base-cased')
```

Instantiating BERT model

```python
model = BertModel.from_pretrained('bert-base-cased', output_hidden_states = False)
model.eval()
```

Evaluating whole dataset

```python
data_dict = { 'word_embedding': [], 'word_label': []}
data_dict_pooler  = { 'pooler_output': [], 'word_label': []}
error_index = []

for i in tqdm(range(len(alldata))):
    text = alldata.iloc[i][0]

    #使用tokenizer將文字轉換成bert輸入需要的 input_ids、 token_type_ids
    inputs = tokenizer.encode_plus(text, return_tensors='pt', add_special_tokens=True)

    input_ids = inputs['input_ids'][:,:512]

    token_type_ids = inputs['token_type_ids'][:,:512]
    
    attention_mask = inputs['attention_mask'][:,:512]

    with torch.no_grad():
        outputs = model(input_ids=input_ids ,attention_mask=attention_mask, token_type_ids=token_type_ids) #https://huggingface.co/transformers/model_doc/bert.html#bertmodel

    token_vecs_last = outputs[0][0][0]
    data_dict['word_embedding'].append(token_vecs_last.numpy())
    data_dict['word_label'].append(alldata.iloc[i][1])
    
    pooler_output = outputs[1][0]
    data_dict_pooler['pooler_output'].append(pooler_output.numpy())
    data_dict_pooler['word_label'].append(alldata.iloc[i][1])    
```

### output result

```python
output_table = pd.DataFrame(data_dict)
output_table.to_pickle("./pickles/output_table4.pkl")
output_table
```

```python
output_table_pooler = pd.DataFrame(data_dict_pooler)
output_table_pooler.to_pickle("./pickles/output_table_pooler4.pkl")
output_table_pooler
```
